# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
x = data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)

data["Clusters"] = kmeans.fit_predict(x)
```

```python
plt.scatter(data["sepal_length"], data["sepal_width"], c=data["Clusters"], cmap="rainbow")
plt.xlabel("Sepal Width", fontsize=15)
plt.ylabel("Sepal Length", fontsize=15)

plt.show()
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

# Clustering (scaled data)

```python
kmeans_scaled = KMeans(3)
kmeans_scaled.fit(x_scaled)
```

```python
cluster_scaled = x.copy()
cluster_scaled["Cluster"] = kmeans_scaled.fit_predict(x_scaled)
cluster_scaled
```

```python
plt.scatter(cluster_scaled["sepal_length"], cluster_scaled["sepal_width"], c=cluster_scaled["Cluster"], cmap="rainbow")
plt.show()
```

```python
wcss=[]

cluster_number = 10
```

## Take Advantage of the Elbow Method

### WCSS

```python
for i in range(1,cluster_number):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
    
wcss
```

### The Elbow Method

```python
number_cluster = range(1,cluster_number)
plt.plot(number_cluster, wcss)

plt.show()
```

Based on the graph, the best number of clusters would be 2 or 3.

K = 2

```python
kmeans_scaled_new = KMeans(2)
kmeans_scaled_new.fit(x_scaled)

cluster_scaled_new = x.copy()
cluster_scaled_new["Cluster"] = kmeans_scaled_new.fit_predict(x_scaled)
cluster_scaled_new

plt.scatter(cluster_scaled_new["sepal_length"], cluster_scaled_new["sepal_width"], c=cluster_scaled_new["Cluster"], cmap="rainbow")
plt.show()
```
